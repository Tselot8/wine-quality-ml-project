```python
import pandas as pd

data = pd.read_csv("../data/winequality-red.csv", sep=';')

# Feature engineering: binary target
data['quality_label'] = data['quality'].apply(lambda x: 1 if x > 5 else 0)

# Drop original target
data.drop('quality', axis=1, inplace=True)

# Save preprocessed dataset
data.to_csv("../data/winequality_processed.csv", index=False)
```

## Data Preprocessing and Feature Engineering

### Objective
The goal of this notebook is to prepare the dataset for machine learning by cleaning and transforming the data.

### Key Tasks Performed
- Converted the original multi-class target variable into a binary classification label
- Removed the original quality score to avoid data leakage
- Saved the processed dataset for use in later stages

### Feature Engineering
- Wines with quality > 5 were labeled as "Good"
